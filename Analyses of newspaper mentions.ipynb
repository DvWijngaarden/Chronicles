## Vergelijkingen tussen kranten en kronieken

```python
corpus = []


def extract_title(path):
    title = os.path.basename(path)
    title = re.sub( r'[.]txt$' , '' , title )
    return title



dir = 'LeeuwarderCourant'


for file in os.listdir(dir):
    if not(re.search(r'^\.' , file)): 
        path = os.path.join(dir,file)
        corpus.append(path)


all_frequencies = dict()
all_words = Counter()
all_lengths = dict()
titles = []

for book in corpus:
    print(book)
    row = []
    
    ## Extract the title from the filename
    title = extract_title(book)
    titles.append(title)
    
    ## Read the full text
    file = open(book,encoding='utf-8')
    full_text = file.read()
    
    ## Find all the words in the text
    words = tokenise_remove_stopwords(full_text.lower())
    all_lengths[title] = len(words)
    
    ## Count all the words in the text
    freq_book = Counter(words)
    all_frequencies[title] = freq_book

    ## Update the Counter object named 'all_words'
    all_words.update(words)
        
print('Done!')

```

```python
import pandas as pd
from collections import Counter

all_rows = []
for book in all_frequencies:
    row = {}
    freq = all_frequencies[book]
    length = all_lengths[book]
    for w in all_words.most_common(20000):
        if length != 0:
            row[w] = freq.get(w, 0) / length
        else:
            row[w] = 0
    all_rows.append(row)

dtm = pd.DataFrame(all_rows, columns=all_words.most_common(20000))
```
